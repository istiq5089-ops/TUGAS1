## **1. Import Library**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## **Penjelasan:**
- **NumPy (np)** digunakan untuk melakukan perhitungan numerik dan menjalankan simulasi distribusi binomial.
- **Matplotlib (plt)** digunakan untuk membuat grafik atau histogram dari hasil simulasi.
- **Pandas (pd)** digunakan untuk membuat dan menampilkan data dalam bentuk tabel agar lebih mudah dibaca.

Library ini penting karena membantu proses analisis data dan visualisasi hasil simulasi.

## **2. Menentukan Parameter**

```python
n = 1000
p = 0.02
hari = 30
```

**Penjelasan:**

Bagian ini digunakan untuk menentukan parameter dalam simulasi distribusi binomial, yaitu:

- **n = 1000**
Menunjukkan jumlah barang yang diproduksi setiap hari.
- **p = 0.02**
Menunjukkan probabilitas atau peluang sebuah barang mengalami cacat produksi yaitu **2%.**
- **hari = 30**
Menunjukkan jumlah hari yang akan disimulasikan yaitu 30 hari ke depan.

Parameter ini akan digunakan untuk mensimulasikan berapa banyak barang cacat yang mungkin terjadi setiap hari.

## **3. Simulasi Distribusi Binomial**

```python
cacat = np.random.binomial(n, p, hari)

data = pd.DataFrame({
    "Hari": range(1, hari+1),
    "Jumlah Cacat": cacat
})

data
```

**Penjelasan:**

Kode ini digunakan untuk melakukan **simulasi jumlah barang cacat menggunakan distribusi binomial.**
- **np.random.binomial(n, p, hari)**
Digunakan untuk menghasilkan angka acak berdasarkan distribusi binomial dengan:
- jumlah percobaan **n = 1000**
- probabilitas cacat **p = 0.02**
- dilakukan selama** 30 hari**

Hasilnya adalah jumlah barang cacat untuk setiap hari.

Kemudian data tersebut dimasukkan ke dalam DataFrame Pandas sehingga dapat ditampilkan dalam bentuk tabel dengan dua kolom:

- **Hari**
- **jumlah cacat**

## **4. Membuat Histogram**

```python
plt.figure(figsize=(8,5))
plt.hist(cacat, bins=10, edgecolor='black')

plt.title("Histogram Jumlah Barang Cacat Selama 30 Hari")
plt.xlabel("Jumlah Barang Cacat")
plt.ylabel("Frekuensi")

plt.show()
```

**Penjelasan:**

Kode ini digunakan untuk membuat **histogram dari hasil simulasi jumlah barang cacat.**

Histogram menunjukkan **distribusi frekuensi jumlah barang cacat selama 30 hari.**

Penjelasan bagian kode:

- plt.figure(figsize=(8,5))
Mengatur ukuran grafik.
- plt.hist()
Membuat histogram dari data jumlah cacat.
- bins = 10
Membagi data menjadi 10 kelompok interval.
- edgecolor='black'
Memberi garis tepi pada histogram agar lebih jelas.
- title, xlabel, ylabel
Memberikan judul grafik serta label pada sumbu X dan Y.

Histogram ini membantu melihat **sebaran jumlah barang cacat yang terjadi selama simulasi.**

## **5. Menghitung Rata-rata dan Standar Deviasi**

```python
rata_rata = np.mean(cacat)
std_dev = np.std(cacat)

print("Rata-rata barang cacat per hari:", rata_rata)
print("Standar deviasi:", std_dev)
```

**Penjelasan:**

Bagian ini digunakan untuk menghitung **statistik dasar dari hasil simulasi.**

- **np.mean(cacat)**
Menghitung** rata-rata jumlah barang cacat per hari.**
- **np.std(cacat)**
Menghitung **standar deviasi**, yaitu ukuran seberapa besar variasi atau penyebaran jumlah barang cacat dari nilai rata-rata.

Nilai ini membantu memahami apakah jumlah barang cacat **relatif stabil atau sangat bervariasi.**

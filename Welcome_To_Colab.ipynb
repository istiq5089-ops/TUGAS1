1. Menampilkan Jumlah Data Hilang

```
import pandas as pd
import numpy as np
import os

file_path = "/dataset_kendaraan.csv"

# Check if the file exists before attempting to read
if os.path.exists(file_path):
    # membaca dataset
    df = pd.read_csv(file_path)

    # menampilkan jumlah missing value setiap kolom
    missing_values = df.isnull().sum()

    print(missing_values)
else:
    print(f"Error: The file '{file_path}' was not found.")
    print("Please make sure the file is uploaded to the Colab environment or provide the correct path.")
    print("You can upload files by clicking on the folder icon on the left sidebar, then the 'Upload to session storage' icon.")
```

2. Mengganti Missing Value pada Variabel Kategori

```
kategori = ["make","bahan-bakar","jumlah-pintu","body-style"]

for col in kategori:
    df[col].fillna("unknown", inplace=True)
```

```
for col in kategori:
    print("Kategori pada kolom:", col)
    print(df[col].value_counts())
    print()
```

3. Mengganti Missing Value pada Variabel Numerik

```
numerik = ["length","width","height"]

for col in numerik:
    mean_value = df[col].mean()
    df[col].fillna(mean_value, inplace=True)
```

4.Min-Max Normalization

```
df["wheel-bases-normalized"] = (
    (df["wheel-bases"] - df["wheel-bases"].min()) /
    (df["wheel-bases"].max() - df["wheel-bases"].min())
)

df.head()
```

5. Z-Score Normalization

```
for col in numerik:
    df[col + "_zscore"] = (df[col] - df[col].mean()) / df[col].std()

df.head()
```

```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv("/data_tugas 2.csv")
df.head()
```

Menampilkan Jumlah Data Hilang

```
missing_values = df.isnull().sum()
print(missing_values)
```

Mengganti Missing Value pada Variabel Kategori

```
kategori = ["make","bahan-bakar","jumlah-pintu","body-style"]

for col in kategori:
    df[col].fillna("unknown", inplace=True)
```

```
for col in kategori:
    print("Ringkasan kategori pada kolom:", col)
    print(df[col].value_counts())
    print()
```

Mengganti Missing Value pada Variabel Numerik

```
numerik = ["length","width","height"]

for col in numerik:
    mean_value = df[col].mean()
    df[col].fillna(mean_value, inplace=True)
```

Min-Max Normalization

```
df["wheel-bases-normalized"] = (
    (df["wheel-bases"] - df["wheel-bases"].min()) /
    (df["wheel-bases"].max() - df["wheel-bases"].min())
)
```

```
display(df.head())
```

Z-Score Normalization

```
for col in numerik:
    df[col + "_zscore"] = (df[col] - df[col].mean()) / df[col].std()
```
